```python
class MyList:
  def __init__(self) -> None:
    self.size=1
    self.n=0

    self.A = self.__make_array(self.size)

  def __len__(self):
    return self.n

  def __resize(self):
    self.size = self.size*2
    arr = self.__make_array(self.size)
    for i in range(self.n):
      arr[i]=self.A[i]
    self.A = arr

  def append(self,item):
    if self.n == self.size:
      #resize
      self.__resize()

    #append
    self.A[self.n] = item
    self.n = self.n + 1

  def __getitem__(self,index):
    if 0 <= index < self.n:
      return self.A[index]
    else:
      return 'IndexOutOfBound'

  def insert(self, pos, item):
    if self.n == self.size:
      self.__resize()

    for i in range(self.n,pos,-1):
      self.A[i] = self.A[i-1]
    self.A[pos]=item
    self.n = self.n + 1

  def find(self,item):
    for i in range(self.n):
      if self.A[i]==item:
        return i
    return 'ValueError'

  def remove(self,item):
    index = self.find(item)
    if type(index)==int:
      self.__delitem__(index)
    else:
      return index


  def pop(self):
    if self.n == 0:
      return 'Empty List'
    print(self.A[self.n-1])
    self.n = self.n - 1

  def __delitem__(self,pos):
    if 0 <= pos < self.n :
      self.n = self.n - 1
      for i in range(pos,self.n):
        self.A[i]=self.A[i+1]


  def __make_array(self, capacity):
    return (capacity*ctypes.py_object)()

  def __str__(self):
    result = ''
    for i in range(self.n):
      result = result + str(self.A[i]) + ','
    return '[' + result[:-1] + ']'
```

```python
#LinkedList

class Node:

  def __init__(self,value) -> None:
    self.data = value
    self.next = None

```

```python
class LinkedList:

  def __init__(self):
    self.head = None
    self.tail = None
    self.n = 0

  def __len__(self):
    return self.n



  def append(self,value):
    node = Node(value)
    if self.tail is None:
      self.tail = node
      self.head = node
      return

    self.tail.next = node
    self.tail = node
    self.n = self.n + 1

  def insert_head(self,value):
    node = Node(value)
    if self.tail is None:
      self.tail = node
      self.head = node
      return

    node.next = self.head
    self.head = node

    self.n = self.n + 1

  def __str__(self):
    curr = self.head
    result = ''
    while curr != None:
      result = result + str(curr.data) + '->'
      curr =  curr.next
    return result[:-2]
```
